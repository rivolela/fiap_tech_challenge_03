```python
import pandas as pd
import requests
import re
from io import StringIO
```

```python

# Replace with the raw GitHub URL to your CSV file
file_path = "https://raw.githubusercontent.com/rivolela/fiap_tech_challenge_03/5074c858801609dc31a8778108950e2e0e3c946f/csv/3_2_taxa_Inflacao.csv"

# Path to your CSV file
#file_path = "../../csv/3_2_taxa_inflacao.csv"

# Fetch the CSV content from the GitHub URL with SSL verification
response = requests.get(file_path, verify=True)

# Decode the content and load it into a pandas DataFrame
csv_content = response.content.decode('ISO-8859-1')

# Load the CSV file with a specified encoding and delimiter, skipping the first row
df = pd.read_csv(StringIO(csv_content), delimiter=';', skiprows=0)
#df = pd.read_csv(file_path, encoding='ISO-8859-1', delimiter=';', skiprows=0)

# Apply function to clean column names
df.columns = [remove_special_characters(col) for col in df.columns]

# Function to remove special characters
def remove_special_characters(text):
    return re.sub(r'[^a-zA-Z0-9\s]', '', text) if isinstance(text, str) else text

# Rename columns with correct names
df.columns = ['Serie', 'Designacao da serie PT', 'Metrica', 'Unidade de Medida', 'Periodo de referencia', 'Valor', 'Estado']


# Apply function to all columns
df = df.map(remove_special_characters)

# Display the first few rows of the DataFrame
print(df.head())


```
